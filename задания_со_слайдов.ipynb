### Задание D

```python
class Parrot:
    def __init__(self, text):
        self.text = text
    
    def say(self, repeat=1):
        for i in range(repeat):
            print(self.text)
    
    def newText(self, text):
        self.text = text

# смотрим
p = Parrot("Гав!")
p.say()           
p.newText("Мяу!")
p.say(3)          
```

### Задание E

```python
import random

class Parrot:
    def __init__(self, text):
        self.phrases = [text]
    
    def say(self, repeat=1):
        phrase = random.choice(self.phrases)
        for i in range(repeat):
            print(phrase)
    
    def learn(self, text):
        self.phrases.append(text)

# смотрим
p = Parrot("Гав!")
p.say()              
p.learn("Мяу!")
p.say()              
p.say(3)             
```

### Задание А

```python
class LampRow:
    def __init__(self):
        self.__state = "00000000"
    
    @property
    def state(self):
        return self.__state
    
    @state.setter
    def state(self, value):
        if len(value) == 8 and all(c in "01" for c in value):
            self.__state = value
        else:
            self.__state = "00000000"
    
    def show(self):
        for c in self.__state:
            print("*" if c == "1" else "-", end="")
        print()

# смотрим
lamps = LampRow()
lamps.show()                    
lamps.state = "10101010"
print(lamps.state)             
lamps.show()                   
```

### Задание B

```python
class LampRow:
    def __init__(self, count=8):
        self.count = count
        self.__state = "0" * count
    
    @property
    def state(self):
        return self.__state
    
    @state.setter
    def state(self, value):
        if len(value) == self.count and all(c in "01" for c in value):
            self.__state = value
        else:
            self.__state = "0" * self.count
    
    def show(self):
        for c in self.__state:
            print("*" if c == "1" else "-", end="")
        print()

# смотрим
lamps = LampRow(6)
lamps.show()                    
lamps.state = "101010"
print(lamps.state)              
lamps.show()                   
lamps.state = "10101010"       
print(lamps.state)             
lamps.show()                    
```

### Задание С

```python
class LampRow:
    def __init__(self, count=8):
        self.count = count
        self.__state = "0" * count
    
    @property
    def state(self):
        return self.__state
    
    @state.setter
    def state(self, value):
        if len(value) == self.count and all(c in "012" for c in value):
            self.__state = value
        else:
            self.__state = "0" * self.count
    
    def show(self):
        for c in self.__state:
            if c == "1":
                print("*", end="")
            elif c == "2":
                print("o", end="")
            else:
                print("-", end="")
        print()

# смотрим
lamps = LampRow(6)
lamps.show()                    
lamps.state = "102102"
print(lamps.state)              
lamps.show()                    
lamps.state = "10201010"        
print(lamps.state)              
lamps.show()                    
```
